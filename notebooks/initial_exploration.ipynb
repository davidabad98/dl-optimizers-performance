### Verify Dataset Structure by Visualizing Sample Images

```python
# Load data
train_loader, test_loader = get_kmnist_dataloaders(batch_size=64)

# Get a single batch of images and labels
images, labels = next(iter(train_loader))

# Convert images to NumPy format for visualization
images = images.numpy()  # Convert PyTorch tensor to NumPy array
```

```python
# Function to show images
def show_images(images, labels, num_images=6):
    fig, axes = plt.subplots(1, num_images, figsize=(12, 6))
    for i in range(num_images):
        ax = axes[i]
        img = images[i].squeeze()  # Remove single channel dimension
        ax.imshow(img, cmap="gray")  # Display as grayscale
        ax.set_title(f"Label: {labels[i].item()}")
        ax.axis("off")
    plt.show()

# Display first few images
show_images(images, labels)
```
